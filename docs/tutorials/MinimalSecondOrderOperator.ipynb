## Operator

This tutorial applies the one-loop counterterm machinery to the minimal second-order operator `D_i^j = delta_i^j Box + W_i^j`.


```
var inputs = new[]
{
    "iK_a^b = d^b_a",
    "K^lm_a^b = d^b_a*g^{lm}",
    "S = 0",
    "W_a^b = W_a^b",
    "F_lmab = F_lmab"
};

inputs
```

## Workflow

The original code sets the Riemann symmetries, defines the `iK`, `K`, `S`, `W`, and `F` tensors, calls `oneloopdiv2(...)`, and then simplifies the resulting counterterms with `EliminateDueSymmetries`.


```
var workflow = new[]
{
    "addSymmetries R_abcd",
    "setSymmetric R_ab",
    "div = oneloopdiv2(iK, K, S, W, F)",
    "counterterms = EliminateDueSymmetries >> div.counterterms"
};

workflow
```

## Output Shape

The final expression is the standard second-order result built from `R^2`, `R_ab R^ab`, bundle curvature `F^2`, and the potential `W` and `Tr(W)` terms.
