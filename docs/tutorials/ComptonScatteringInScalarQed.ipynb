## Theory

The original tutorial derives the scalar-QED Compton amplitude from three diagrams and then computes the polarization-summed squared matrix element in Mandelstam variables.


```
var rules = new[]
{
    "D(k) = -i/(k^2 - m^2)",
    "V_mu(k, p) = -i*e*(k_mu + p_mu)",
    "V_munu = 2*i*e^2*g_munu"
};

rules
```

## Workflow

The original code builds the three diagram contributions, substitutes the vertex and propagator rules, forms `M_ij*M^ij/2`, applies Mandelstam substitutions, expands, eliminates metrics, sets `d_i^i = 4`, and factors the result.


```
var workflow = new[]
{
    "V_i[p_a, q_b] = -I*e*(p_i+q_i)",
    "V_ij = 2*I*e**2*g_ij",
    "D[k_a] = -I/(k^a*k_a - m**2)",
    "setMandelstam([p1_a: 0, k1_a: m, p2_a: 0, k2_a: m])",
    "ExpandAll & EliminateMetrics & mandelstam",
    "Factor"
};

workflow
```

```
var matrixElement = TensorFactory.Parse("M^ij");

new
{
    MatrixElementSymbol = matrixElement.ToString(),
    SquaredMatrixElementConcept = "M_ij*M^ij/2",
    FinalVariables = new[] { "s", "t", "u", "m", "e" }
}
```
