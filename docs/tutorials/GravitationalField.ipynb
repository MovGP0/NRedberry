## Problem

This tutorial evaluates one-loop counterterms for the gravitational field in the `lambda`-family gauges, including the gauge-fixed graviton operator and the separate Faddeev-Popov ghost contribution.


```
var ingredients = new[]
{
    "iK_ab^cd for the graviton field",
    "K^lm_ab^cd",
    "S = 0",
    "W_ab^cd with P and curvature terms",
    "F_lm^kd_pr = R^k_plm*d^d_r + R^d_rlm*d^k_p"
};

ingredients
```

## Workflow

The original code sets the Riemann symmetries, defines an auxiliary tensor `P_cd^lm`, substitutes it into `W`, runs `oneloopdiv2(iK, K, S, W, F)`, and then expands and factors the resulting counterterms.


```
var workflow = new[]
{
    "addSymmetries R_abcd",
    "setSymmetric R_ab",
    "W <<= P",
    "oneloopdiv2(iK, K, S, W, F).counterterms",
    "ExpandAll & Factor[[FactorScalars: false]]"
};

workflow
```

## Output Shape

The graviton contribution is expressed in terms of `R_ab R^ab` and `R^2`, after which the tutorial subtracts twice the ghost contribution from the minimal second-order operator to obtain the full gauge-fixed result.
