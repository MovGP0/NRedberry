## Theory

This tutorial upgrades the scalar example to spinor QED. The amplitude uses two diagrams, electron propagators, Dirac matrices, external spinors, and photon polarization vectors.


```
var ingredients = new[]
{
    "electron propagator D(k)",
    "vertex V_mu = -i*e*gamma_mu",
    "external spinors u, ubar",
    "polarization vectors e_mu(p)"
};

ingredients
```

## Workflow

The original code defines the matrix objects, substitutes the Feynman rules, simplifies with Mandelstam identities, constructs the conjugate amplitude with reversed gamma-matrix order, sums polarizations, takes Dirac traces, and then factors the final Klein-Nishina-Tamm expression.


```
var workflow = new[]
{
    "defineMatrices for gamma, vertex, propagator, spinors",
    "M = Ma + Mb",
    "Conjugate & Reverse[Matrix1]",
    "DiracTrace[Gamma: G_a]",
    "ExpandAndEliminate & mandelstam",
    "Factor"
};

workflow
```

```
var polarizationSums = new[]
{
    "e_m[p1_a]*e_n[p1_a] = -g_mn",
    "e_m[p2_a]*e_n[p2_a] = -g_mn",
    "vu[k2_m]*cu[k2_m] = m + k2^m*G_m",
    "vu[k1_m]*cu[k1_m] = m + k1^m*G_m"
};

polarizationSums
```
