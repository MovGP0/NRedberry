## Process

This tutorial computes the leading-order squared matrix element for `e+ e- -> mu+ mu-`. The setup is structurally simpler than the Compton examples because there is only one tree-level diagram.


```
var ingredients = new[]
{
    "gamma matrices",
    "QED vertex",
    "photon propagator",
    "electron and muon spinors"
};

ingredients
```

## Workflow

The original code substitutes the photon propagator and vertex, applies Mandelstam and mass-shell substitutions, constructs the conjugate amplitude, sums over fermion polarizations, evaluates the Dirac trace, and factors the result.


```
var workflow = new[]
{
    "G_mn[k_a] = -I*g_mn/(k_a*k^a)",
    "V_i = -I*e*G_i",
    "setMandelstam([p1_m: me, p2_m: me, k1_m: mu, k2_m: mu])",
    "DiracTrace[Gamma: G_a]",
    "ExpandAndEliminate & mandelstam",
    "Factor"
};

workflow
```
