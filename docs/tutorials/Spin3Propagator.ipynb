## Problem

This tutorial solves for the spin-3 propagator by inverting a fully symmetrized quadratic form in momentum space. The unknown propagator tensor `P^{abc}_{pqr}` is determined from a linear tensor equation solved with `Reduce`.


```
var ingredients = new[]
{
    "quadratic form K with parameters l and b",
    "upper-index symmetrization",
    "lower-index symmetrization",
    "propagator tensor P^{abc}_{mnk} with full upper/lower symmetries"
};

ingredients
```

## Workflow

The original code defines the quadratic form `K`, symmetrizes its upper and lower indices with `Symmetrize`, declares the symmetry generators of the unknown propagator, forms the tensor equation `K^abc_mnk*P^mnk_pqr = d^a_p*d^b_q*d^c_r`, and then solves it with `Reduce` under the assumptions `d_n^n = 4` and `n_a*n^a = 1`.


```
var workflow = new[]
{
    "K = ...",
    "Symmetrize[upper] >> K",
    "Symmetrize[lower] >> K",
    "addSymmetries P^{abc}_{mnk}",
    "lhs = K^abc_mnk*P^mnk_pqr",
    "rhs = d^a_p*d^b_q*d^c_r",
    "Reduce([lhs = rhs], [P^{abc}_{pqr}], options)"
};

workflow
```

## Result Structure

The original page reports the full propagator in a symmetrized basis built from Kronecker deltas, metrics, and momentum vectors, with coefficients `c0` through `c9` depending on `lambda` and `beta`. The explicit expression is large, but its structure is the main point: `Reduce` can solve this highly constrained tensor system directly once the symmetries are declared.


```
var resultShape = new[]
{
    "Sym[c0 * n_b1 n_b2 n_b3 n^a1 n^a2 n^a3 + ...]",
    "coefficients c0 ... c9 as functions of l and b",
    "dimension fixed to 4 with n^2 = 1 during the solve"
};

resultShape
```
