## Problem

This tutorial treats the squared vector-field operator, a fourth-order construction whose counterterms are expected to be exactly twice the vector-field result because `Tr ln D^2 = 2 Tr ln D`.


```
var ingredients = new[]
{
    "four-index K tensor",
    "W^{lm}_a^b with curvature and P terms",
    "M_a^b with curvature-squared and P terms",
    "iK_a^b = d_a^b + (2*g + g**2) n_a n^b",
    "F_lmab = R_lmab"
};

ingredients
```

## Workflow

The original code defines `K`, `W`, and `M` in terms of `lambda`, substitutes `lambda = g/(1+g)`, runs `oneloopdiv4(...)`, and then factors the final counterterms with `Factor[[FactorScalars: false]]`.


```
var workflow = new[]
{
    "lambda = g/(1+g)",
    "oneloopdiv4(iK, K, S, W, N, M, F)",
    "Factor[[FactorScalars: false]] & P_l^l = P"
};

workflow
```

## Output Shape

The final expression has the same invariant structure as the vector-field case but with coefficients doubled, exactly as expected from the operator identity explained in the tutorial.
