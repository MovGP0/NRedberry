## Scope

The original page's theory section is abbreviated, but the code shows a realistic multi-diagram weak-decay calculation with Dirac traces, Levi-Civita simplification, propagators, spin projection for the `B_c` state, and a final Mathematica export.


```
var buildingBlocks = new[]
{
    "DiracTrace with G_a, G5, e_abcd",
    "LeviCivitaSimplify.minkowski",
    "spinor propagator D[m, q_a]",
    "spin projection b[p1]*c[p2] = F*(p*G + M)*G5",
    "mass-shell and scalar-product substitutions" 
};

buildingBlocks
```

## Workflow

The original code constructs four emission diagrams, substitutes the propagator and spin projection, reduces the expression with mass and scalar-product identities, conjugates with reversed matrix order and a `G5` sign flip, sums spins and photon polarizations, evaluates traces, enforces momentum conservation, and finally exports the result in Mathematica format.


```
var workflow = new[]
{
    "Ma + Mb + Mf + Me",
    "spinProjection & propagator",
    "ExpandAll & EliminateMetrics & scalars",
    "Conjugate & Reverse[Matrix1] & G5 = -G5",
    "DiracTrace",
    "LeviCivitaSimplify.minkowski",
    "p_a = k1_a + k2_a + k3_a",
    "OutputFormat.WolframMathematica"
};

workflow
```
