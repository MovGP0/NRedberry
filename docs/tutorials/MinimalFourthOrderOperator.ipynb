## Operator

This notebook covers the minimal fourth-order operator `D_i^j = delta_i^j Box^2 + W^{mu nu}_i^j nabla_mu nabla_nu + M_i^j`.


```
var inputs = new[]
{
    "iK_a^b = d_a^b",
    "K^lmcd_a^b = d_a^b*1/3*(...)",
    "S = 0",
    "W^lm_a^b = W^lm_a^b",
    "N = 0",
    "M_a^b = M_a^b",
    "F_lmab = F_lmab"
};

inputs
```

## Workflow

The original code defines the fourth-order inputs, calls `oneloopdiv4(...)`, then introduces shorthand substitutions like `M^l_l = M` and `W_lm^a_a = W_lm` before printing the result.


```
var workflow = new[]
{
    "addSymmetries R_abcd",
    "addSymmetry W_lmab on indices [0,1]",
    "div = oneloopdiv4(iK, K, S, W, N, M, F)",
    "counterterms = EliminateDueSymmetries >> div.counterterms"
};

workflow
```

## Output Shape

The final answer combines curvature-squared terms, bundle curvature `F^2`, traces of `W`, and the scalar `M`, matching the standard fourth-order minimal-operator result.
