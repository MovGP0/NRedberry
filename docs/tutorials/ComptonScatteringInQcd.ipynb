## Theory

The QCD tutorial adds SU(N) color structure, gluon polarization sums in axial gauge, the three-gluon vertex, and both Dirac and unitary traces. The final result must be independent of the auxiliary gauge vector `n_mu`.


```
var ingredients = new[]
{
    "quark propagator",
    "gluon propagator in axial gauge",
    "quark-gluon vertex",
    "three-gluon vertex",
    "Dirac matrices and SU(N) matrices"
};

ingredients
```

## Workflow

The original code defines Matrix1 and Matrix2 objects, sets symmetric `d_ABC` and antisymmetric `f_ABC`, substitutes all QCD Feynman rules, simplifies with Mandelstam identities and `n_a*n^a = 1`, constructs the conjugate amplitude with reversed matrix order, performs Dirac and unitary traces, and finally exports the large expression in Mathematica syntax for final factorization.


```
var workflow = new[]
{
    "defineMatrices for T_A, G_a, quark spinors, vertices, propagator",
    "setSymmetric d_ABC",
    "setAntiSymmetric f_ABC",
    "Reverse[Matrix1, Matrix2]",
    "DiracTrace",
    "UnitaryTrace",
    "UnitarySimplify",
    "OutputFormat.WolframMathematica"
};

workflow
```
