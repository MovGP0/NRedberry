## Problem

This tutorial evaluates the one-loop counterterms for the massive vector-field operator in the `lambda`-family gauge and rewrites the result in terms of `gamma = lambda / (1 - lambda)`.


```
var inputs = new[]
{
    "iK_a^b = d_a^b + g*n_a*n^b",
    "K^{mn}_i^j = g^{mn} d_i^j - ...",
    "S = 0",
    "W^a_b = P^a_b + g/(2*(1+g))*R^a_b",
    "F_abcd = R_abcd"
};

inputs
```

## Workflow

The original code computes `div = oneloopdiv2(iK, K, S, W, F)`, extracts `div.counterterms[1]`, and then uses `Collect` and `Factor` to organize the result by `R` and `P` invariants.


```
var workflow = new[]
{
    "addSymmetries R_abcd",
    "setSymmetric R_ab, P_ab",
    "oneloopdiv2(iK, K, S, W, F)",
    "Collect[ R, P, Factor ]"
};

workflow
```

## Output Shape

The resulting counterterms involve `R_ab R^ab`, `R^2`, `P_ab P^ab`, `P^2`, `R_ab P^ab`, and `R*P`, with coefficients depending on the gauge parameter.
