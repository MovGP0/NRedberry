## Description

- `Symmetrize[indices]` gives an expression the symmetries carried by `indices`.
- It acts only with respect to the specified indices, not necessarily all indices of the expression.
- It multiplies by the appropriate symmetry factor.


```
var examples = new[]
{
    "Symmetrize _ab on T_ab",
    "AntiSymmetrize _ab on T_ab",
    "Symmetrize _abc on T_abcde",
    "Use a custom permutation group on _abcd"
};

examples
```

## Examples

The original page shows ordinary symmetrization, antisymmetrization, partial-index symmetrization, and applying a full custom symmetry group carried by a `SimpleIndices` object.


```
var sample = TensorFactory.Parse("T_ac*F^c_b - F_bc*T^c_a");

new
{
    Input = sample.ToString(),
    SymmetrizeConcept = "Symmetrize[indices] >> expr",
    FullSymmetricConcept = "indices = expr.indices.si; indices.symmetries.setSymmetric()"
}
```

## Details

Symmetries are defined relative to the supplied index ordering. The same permutation group applied to `_abc` and `_cab` produces different transformed expressions because the reference ordering differs.
