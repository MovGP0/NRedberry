## Basics

`SimpleTensor` represents simple tensor symbols such as `f_mn` or scalar symbols such as `x`. Like every other expression kind, it inherits the common tensor properties, but it has two distinctive traits:

- its indices are `SimpleIndices`
- it is the place where named tensor symmetries are attached and reused


```
var tensor = TensorFactory.Parse("f_cba^\\alpha_AB");

new
{
    Tensor = tensor.ToString(),
    Indices = tensor.Indices.ToString(),
    SimpleTensorConvention = "indices sorted by type, preserving relative order within each type"
}
```

## Symmetries

The original page emphasizes that symmetries belong to simple tensors. Once attached to one representative such as `f_pqr`, they apply automatically to all tensors with the same symbolic name and the same index structure.


```
var tensor = TensorFactory.ParseSimple("f_pqr");
int[] symmetry = [1, 0, 2];
tensor.SimpleIndices.Symmetries.AddSymmetry(symmetry.p);

var sameFamily = TensorFactory.ParseSimple("f_klm");

new
{
    Tensor = tensor.ToString(),
    RegisteredGenerators = tensor.SimpleIndices.Symmetries.Generators.Count,
    SameFamilyOrder = sameFamily.SimpleIndices.Symmetries.PermutationGroup.Order
}
```

## Details

The textual name of a simple tensor is available as `stringName`, but the mathematical identity is carried by its encoded `name`, which depends on both the symbol and the index structure.


```
var comparisons = new[]
{
    "f_ab and f_bc share the same encoded name",
    "f_ab and t_bc do not",
    "f_ab and f_b do not",
    "f_ab and f_AB do not"
};

comparisons
```
