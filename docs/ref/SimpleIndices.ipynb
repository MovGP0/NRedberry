## Basics

`SimpleIndices` represents the indices of `SimpleTensor` and `TensorField`. Unlike general `Indices`, it preserves the relative ordering of indices exactly as supplied by the user within each index type.


```polyglot-notebook
var orderingExamples = new[]
{
    "_pqrs^sp",
    "_{mn}^{\\beta\\alpha}_{ba\\alpha}"
};

orderingExamples
```

## Features

All the standard index views remain available while preserving ordering: `free`, `inverted`, `upper`, `lower`, indexing by position, slicing, and selecting by index type.


```polyglot-notebook
var featureExamples = new[]
{
    "indices.free",
    "indices.inverted",
    "indices.upper",
    "indices.lower",
    "indices[0]",
    "indices[LatinUpper, 0]",
    "indices[1, 3, 4]",
    "indices[1..4]"
};

featureExamples
```

## Symmetries

`SimpleIndices` has a `Symmetries` property that stores permutation symmetries and exposes the corresponding permutation group. Those symmetries are used by transformations such as `Symmetrize`, as well as tensor generation and solving workflows.


```polyglot-notebook
var fourIndexTensor = TensorFactory.ParseSimple("R_abcd");
int[][] fourIndexSymmetry = [[0, 1], [2, 3]];
fourIndexTensor.SimpleIndices.Symmetries.AddSymmetry(fourIndexSymmetry.p);

var sixIndexTensor = TensorFactory.ParseSimple("T_abcdef");
int[][] sixIndexSymmetry = [[0, 3, 5]];
sixIndexTensor.SimpleIndices.Symmetries.AddSymmetry(sixIndexSymmetry.p);

new
{
    FourIndexOrder = fourIndexTensor.SimpleIndices.Symmetries.PermutationGroup.Order,
    SixIndexOrder = sixIndexTensor.SimpleIndices.Symmetries.PermutationGroup.Order,
    RelatedOperations = new[] { "Symmetrize", "GenerateTensor", "Reduce" }
}
```
