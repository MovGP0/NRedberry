## Basics

`TensorField` represents tensor-valued functions such as `f_mn[x_a]` and tensor-valued derivatives such as `f~(1)_a[x_i]`. It inherits from both `Tensor` and `SimpleTensor`, so it keeps container behavior, naming, and symmetry features while also carrying function arguments.


```
var field = TensorFactory.Parse("f_cba^\\alpha_AB[t_mn, f_i]");

new
{
    Field = field.ToString(),
    Indices = field.Indices.ToString(),
    Convention = "same ordering convention as SimpleTensor, with SimpleIndices"
}
```

## Symmetries

Symmetries of `TensorField` are attached in the same way as for `SimpleTensor`. Once declared for one representative, they apply to all tensor fields with the same symbolic name and compatible index structure.


```
var field = (TensorField)TensorFactory.Parse("f_pqr[k_l, k_l]");
int[] symmetry = [1, 0, 2];
field.SimpleIndices.Symmetries.AddSymmetry(symmetry.p);

var sameFieldFamily = (TensorField)TensorFactory.Parse("f_klm[k_a, k_b]");

new
{
    Field = field.ToString(),
    RegisteredGenerators = field.SimpleIndices.Symmetries.Generators.Count,
    SameFamilyOrder = sameFieldFamily.SimpleIndices.Symmetries.PermutationGroup.Order
}
```

## Derivatives

For derivatives, the total indices are formed by concatenating the parent-field indices with the inverted indices of the differentiating variables, grouped slot by slot. Repeated differentiation with respect to the same slot gives symmetric derivative indices.


```
var derivative = TensorFactory.Parse("F~(1,2,0)_{ab cd mn}[x_mn, t_i, x]");

new
{
    Derivative = derivative.ToString(),
    DerivativeConcepts = new[]
    {
        "isDerivative()",
        "getDerivativeOrder(i)",
        "parentField",
        "partitionOfIndices"
    }
}
```
