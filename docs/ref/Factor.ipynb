## Description

- `Factor` factors a symbolic polynomial over the integers.
- `Factor` acts only on scalar parts of an expression.
- `Factor` acts only on the top algebraic level.
- `Factor[[FactorScalars: false]]` restricts factoring to purely symbolic scalar expressions and avoids treating indexed scalar combinations as independent variables.


```
var examples = new[]
{
    "2*x**3*y - 2*a**2*x*y - 3*a**2*x**2 + 3*a**4",
    "(a+b)**4*F_mn + (x**6-y**6)*R_mn",
    "(a+b)**2*f_m*f^m + (a**2 - b**2)*f_a*f^a*f_b*f^b"
};

examples
```

## Examples

The original page shows three characteristic cases:

- factoring a multivariate polynomial
- factoring only the scalar coefficients in a tensorial expression
- factoring scalar contractions of tensors, unless `FactorScalars: false` is used


```
var sample = TensorFactory.Parse("(a+b)**2*f_m*f^m + (a**2 - b**2)*f_a*f^a*f_b*f^b");

new
{
    Input = sample.ToString(),
    FactorConcept = "Factor >> expr",
    NoTensorScalarsConcept = "Factor[[FactorScalars: false]] >> expr"
}
```
