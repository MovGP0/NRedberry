## Description

- `Reduce(equations, vars)` reduces a system of tensorial equations to a symbolic system.
- `Reduce(equations, vars, options)` adds simplification rules and optional external solving.
- With `ExternalSolver`, Redberry can fully solve the resulting symbolic equations in Maple or Mathematica.


```
var examples = new[]
{
    "Reduce([eq1, eq2], [F_mn])",
    "Reduce with ExternalSolver = Mathematica",
    "Reduce with symmetry assumptions on the unknown tensor",
    "Reduce a generated projector ansatz"
};

examples
```

## Examples

The original page covers direct reduction, fully solved systems via Mathematica, solving under symmetry constraints, and solving for generated tensor ansätze with free parameters.


```
var eq1 = TensorFactory.Parse("F_mn + F_nm + A_mn + A_nm = 0");
var eq2 = TensorFactory.Parse("F_mn/2 + A_nm/2 = F_nm + A_mn");

new
{
    Eq1 = eq1.ToString(),
    Eq2 = eq2.ToString(),
    ReduceConcept = "Reduce([eq1, eq2], [\"F_mn\"])"
}
```

## Options

The documented options are:

- `Transformations`: simplifications applied to the generated scalar system.
- `SymmetricForm`: request symmetrized representative solutions.
- `GeneratedParameters`: control the naming of introduced free parameters.
- `ExternalSolver`: configure Maple or Mathematica, including path and free-parameter handling.
