## Basics

- `Transformation` is the common supertype for transformations such as `Expand` or `EliminateMetrics`.
- Transformations apply to expressions with the conceptual `>>` operator or by direct invocation of a transform method.
- Transformations can be composed with `&` to run sequentially.
- `Identity` is the no-op transformation.


```
var examples = new[]
{
    "Expand >> (a+b)*(c+d)",
    "f_a[x_a] = x_a >> g^b[z_a] = f^b[z^i]",
    "Expand & EliminateMetrics",
    "custom transformation from a function"
};

examples
```

## Examples

The original page shows ordinary built-in transformations, transformation-valued equations, composition with `&`, and small custom transformations that are constructed from functions or closures.


```
var sample = TensorFactory.Parse("f_ab^cd");

new
{
    Input = sample.ToString(),
    InvertIndicesConcept = "(expr.indices % expr.indices.inverted) >> expr",
    CombinedConcept = "Expand & EliminateMetrics"
}
```
