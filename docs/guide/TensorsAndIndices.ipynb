## Central Object Types

The original guide treats `Tensor`, `Indices`, and `Transformation` as the three central object families. This notebook section focuses on tensors and indices. Transformations are covered in the later transformations notebooks.


## Types of Tensors

Redberry expressions are all tensors, but they can appear as several concrete shapes: simple tensors, tensor fields, products, sums, powers, numbers, and substitution-like expressions.


```
var expr = TensorFactory.Parse("a*T_mn + g_mn");

new
{
    TypeName = expr.GetType().Name,
    IsTensor = expr is Tensor,
    IsSum = expr is Sum,
    IsProduct = expr is Product
}
```

## Structure of Tensors

A tensor has indices and content. Sums expose summands, products expose multipliers, and tensor fields expose arguments. The container-like shape is useful when inspecting or rebuilding an expression tree.


```
var expression = TensorFactory.Parse("a*F^a_mn + b^a*h_m*h_n + t_mn^a");

var elements = new List<string>();
foreach (var item in expression)
{
    elements.Add(item.ToString());
}

new
{
    Indices = expression.Indices.ToString(),
    Size = expression.Size(),
    First = expression[0].ToString(),
    Third = expression[2].ToString(),
    Elements = elements
}
```

## Modifying Expressions

Tensors are immutable. Operations such as `Set` and `Remove` return new tensors and leave the original expression unchanged.


```
var original = TensorFactory.Parse("A_i + C_i + D_i");
var replaced = original.Set(1, TensorFactory.Parse("B_i"));
var removedOne = original.Remove(0);
var removedRange = original.Remove(0, 2);

new
{
    Original = original.ToString(),
    Replaced = replaced.ToString(),
    RemovedOne = removedOne.ToString(),
    RemovedRange = removedRange.ToString()
}
```

## Working with Indices

The `Indices` object exposes the total, free, inverted, and upper-index views of a tensor expression. The original guide uses these accessors directly, and the same style maps well into the C# port.


```
var indexed = TensorFactory.Parse("2*x_am*f^m*(a^n+b^n)");
var indices = indexed.Indices;

new
{
    All = indices.ToString(),
    Size = indices.Size(),
    Free = indices.Free.ToString(),
    Inverted = indices.Inverted.ToString(),
    Upper = indices.Upper.ToString()
}
```

## Ordering of Indices

Simple tensors preserve the user-supplied order inside each index type, while composite expressions can sort indices according to Redberry's internal rules. This is one of the reasons the index container type differs between simple tensors and composite tensors.


```
var simple = TensorFactory.Parse("F_mn^{\\beta\\alpha}_{ba\\alpha}");
var product = TensorFactory.Parse("F_mn*F^{\\beta\\alpha}*F_ba\\alpha");
var sum = TensorFactory.Parse("R^a_amn^\\alpha + K^i_inm^\\alpha");

new
{
    SimpleIndices = simple.Indices.ToString(),
    ProductIndices = product.Indices.ToString(),
    SumIndices = sum.Indices.ToString()
}
```

## Single Indices and Parsing

The original guide also covers low-level single-index parsing and `SimpleIndices` parsing. At the notebook level, the most common entry point remains tensor parsing plus `tensor.Indices`, but the same concepts still apply: index state, type, inversion, and ordering are first-class parts of the symbolic model.
