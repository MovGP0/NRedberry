## Inputting Expressions

The notation used for symbolic input is close to what the original Redberry guide describes: tensor expressions use a LaTeX-like syntax with a few practical parsing rules.


```
var t1 = TensorFactory.Parse("a * F^{A}_{B \\mu \\nu}");
var t2 = TensorFactory.Parse("a * F^A_B\\mu\\nu");
var t3 = TensorFactory.Parse("F^A_{B_{21}C\\mu\\nu}");
var t4 = TensorFactory.Parse("\\Gamma^A_{\\alpha\\beta}");
var t5 = TensorFactory.Parse("\\Gamma{}^A{}_{\\alpha\\beta}");

new[] { t1, t2, t3, t4, t5 }
```

These examples mirror the original guide:

- braces are often optional for simple runs of indices
- braces are required when spaces or nested subscripts would make the parse ambiguous
- LaTeX-style names such as `\\Gamma` are accepted


### Functions

Functions follow the same symbolic input style as described on the original page, including user-defined tensorial functions and built-in scalar functions.


```
var fieldExpr = TensorFactory.Parse("F^a_bcd[G^a_bc, p^a]");
var builtInExpr = TensorFactory.Parse("Sin[m**2 - p_m*p^m] - Log[x/2]");
var powerExpr = TensorFactory.Parse("Power[a, b]");

new { fieldExpr, builtInExpr, powerExpr }
```

## Printing Expressions and Exporting to Other CASs

The original guide explains that Redberry objects have several output formats. NRedberry uses the same general pattern through `ToString(OutputFormat. ...)`.


```
var expr = TensorFactory.Parse("a * F^{A}_{A m n} + g_mn");
expr.ToString()
```

```
var sample = TensorFactory.Parse("F_mn^{\\alpha\\beta}/(a+b)");

new
{
    Latex = sample.ToString(OutputFormat.LaTeX),
    Cadabra = sample.ToString(OutputFormat.Cadabra),
    Utf8 = sample.ToString(OutputFormat.UTF8),
    Mathematica = sample.ToString(OutputFormat.WolframMathematica),
    Maple = sample.ToString(OutputFormat.Maple)
}
```

By default, notebook output stays in the native Redberry/NRedberry style. Alternate renderings are useful when moving expressions into TeX, Maple, Mathematica, Cadabra, or similar systems.
