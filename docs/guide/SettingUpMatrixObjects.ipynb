## Matrix Indices

Matrix objects are tensors with a subset of indices marked as nonmetric matrix indices. That allows Redberry to model Dirac matrices, SU(N) matrices, spinors, and similar objects without printing the inserted internal matrix indices by default.


```
var basicExamples = new[]
{
    "defineMatrices \"G_a\", Matrix1.matrix",
    "G_a*G_b",
    "Tr[G_a*G_b]"
};

basicExamples
```

## Hidden Matrix Indices

The original guide shows that explicit output formats reveal the inserted matrix indices. This is useful for understanding how the parser turns plain-looking matrix strings into tensors with contracted matrix slots.


```
var explicitPrintConcepts = new[]
{
    "G_a^a'_b'*G_b^b'_c'",
    "G_a^a'_b'*G_b^b'_a'"
};

explicitPrintConcepts
```

## Matrix Signatures

The guide defines matrices, vectors, covectors, and more general matrix tensors by their number of upper and lower matrix indices. The predefined nonmetric index types are `Matrix1` through `Matrix4`.


```
var signatures = new[]
{
    "Matrix1.matrix",
    "Matrix1.vector",
    "Matrix1.covector",
    "Matrix1.tensor(2, 3)"
};

signatures
```

## Traces and Identity Insertion

If a sum mixes matrix-valued and non-matrix-valued terms, Redberry inserts identity matrices as Kronecker deltas so the sum has consistent free matrix indices. Traces can be taken over all matrix types or restricted to one matrix-index family.


```
var traceExamples = new[]
{
    "Tr[A*B]",
    "Tr[A*B, Matrix1]"
};

traceExamples
```
