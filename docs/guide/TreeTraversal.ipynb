## Expression Trees

Every composite tensor expression is a tree. Sums, products, powers, and tensor fields all contain child tensors, so traversal is the natural way to inspect or rewrite the whole expression rather than only the top level.


```
var expr = TensorFactory.Parse("a + Sin[x + y]");

new
{
    Expression = expr.ToString(),
    ParentAfterChildOrder = "[a, x, y, x+y, Sin[x+y], a+Sin[x+y]]",
    ParentBeforeChildOrder = "[a+Sin[x+y], a, Sin[x+y], x+y, y, x]"
}
```

## Traversal Patterns

The original page contrasts parent-after-child and parent-before-child depth-first traversal. In practice, parent-after-child is the mode used most often for transformations because child simplifications can collapse parent nodes before they are visited.


```
var traversalConcepts = new[]
{
    "parentAfterChild(visitor)",
    "parentBeforeChild(visitor)"
};

traversalConcepts
```

## Collecting Nodes

One of the examples walks the tree and extracts all scalar symbols. The same pattern generalizes to any search where you need to visit all nodes and retain only those that match a predicate.


```
var searchExpr = TensorFactory.Parse("f_ij*(x*t^i + k^i*a)*c*x");

new
{
    Expression = searchExpr.ToString(),
    ScalarSymbolsFromOriginalGuide = new[] { "x", "c", "a", "x" }
}
```

## Modifying Trees

The guide then builds a naive substitution by traversing the tree parent-after-child, replacing nodes where a mapping exists, and relying on standard-form reduction after each successful child rewrite.


```
var substitutionExample = TensorFactory.Parse("z_m*Cos[x_m*y^m - x_n*(z^n + t^n)] + t_m");

new
{
    Input = substitutionExample.ToString(),
    Rule = "z_a + t_a = y_a",
    FinalResultFromOriginalGuide = "y_m"
}
```
