## Basics

Substitution is the most common transformation in everyday Redberry work. The defining feature is that dummy-index clashes are repaired automatically during application.


```
var substitution = TensorFactory.Parse("x = x_a^a");
var tensor = TensorFactory.Parse("(x*f_a + y_a)*(x*f_b + z_b)");

new
{
    Substitution = substitution.ToString(),
    Tensor = tensor.ToString(),
    ExpectedResult = "(x_d^d*f_a + y_a)*(x_c^c*f_b + z_b)"
}
```

## Tensor-Field Substitutions

Substitutions can target tensor fields as well as simple tensors. When the arguments are indexed, the matching can become ambiguous, so the original guide allows explicit index-order annotations in the field argument list.


```
var tensorFieldExamples = new[]
{
    "F_ij[x_m, y_m] = x_i*y_j",
    "F_k[x_i*y_j]",
    "F_k[x_i*y_j:_ji]"
};

tensorFieldExamples
```

## Products, Sums, and Symmetry-Aware Matching

Redberry does not limit substitutions to simple leaves. The original guide shows substitutions against sums and products while respecting permutation symmetries and contraction structure.


```
var structuredExamples = new[]
{
    "f_m + R_bma*F^ba - R_ljm*F^lj = R_bam*F^ab",
    "K_a*(A^ab - A^ba) = F^a*A_a^b",
    "Cos[a - b] = c",
    "Sin[a - b] = s"
};

structuredExamples
```

## Sequential vs. Simultaneous Rules

The original guide distinguishes `&` from `|`: `&` applies substitutions sequentially, while `|` applies the rules simultaneously. That distinction matters whenever the rules overlap, such as `x = y` and `y = x`.


```
var compositionExamples = new[]
{
    "x = y & y = x",
    "x = y | y = x",
    "x + 2*y"
};

compositionExamples
```

## Constructing Substitutions Programmatically

The original page also constructs substitutions from previously computed left-hand and right-hand sides. In notebook C#, the same idea is straightforward: build the tensor objects first, then create the substitution expression that will be applied later.


```
var lhs = TensorFactory.Parse("x");
var rhs = TensorFactory.Parse("a + b");

new
{
    Left = lhs.ToString(),
    Right = rhs.ToString(),
    EquivalentSubstitutionSyntax = "x = a + b"
}
```
