## Covariant and Contravariant Indices

Redberry distinguishes lower and upper indices explicitly. Two indices contract only when they have the same name and type but opposite state.


```
var valid = TensorFactory.Parse("F_a^a");
var mixed = TensorFactory.Parse("F_mn*(A^ab + M_m*N^mab)");
var powered = TensorFactory.Parse("F_a*(A^a*B_a)**2");

new
{
    Valid = valid.ToString(),
    SumScopeRenaming = mixed.ToString(),
    PowerScopeRenaming = powered.ToString()
}
```

## Illegal Contractions

The original guide highlights two invalid situations:

- using the same index name twice with the same state in a single tensor slot
- constructing a product where equal-state duplicates make the contraction structure meaningless

Those inputs should raise an index-consistency exception rather than silently producing an invalid tensor.


```
var illegalSamples = new[]
{
    "F_aa",
    "F_ij*M^i*N^j*K^j"
};

illegalSamples
```

## Automatic Dummy-Index Renaming

A key part of the Redberry model is that dummy-index clashes are repaired immediately after parsing. This makes later transformations safer because the invalid intermediate forms never survive in the expression tree.


```
var renamingFromSumScope = TensorFactory.Parse("F_mn*(A^ab + M_m*N^mab)");
var renamingFromPowerScope = TensorFactory.Parse("F_a*(A^a*B_a)**2");

new
{
    FromSumScope = renamingFromSumScope.ToString(),
    FromPowerScope = renamingFromPowerScope.ToString()
}
```
