## Graph-Based Matching

This page explains the architectural choice behind Redberry's tensor comparison engine. Instead of canonicalizing all indices, it treats the contraction pattern of a product as a graph and searches for graph isomorphisms or automorphisms.


## Why Not Canonicalization

The original text contrasts the graph-isomorphism approach with index canonicalization and double-coset enumeration. The argument is pragmatic: canonicalization can be expensive, depends on total index count, and is less natural for free-index relabeling and nested non-product structures.

## Symmetry Detection and Zero Recognition

For expressions that vanish because of conflicting symmetry and antisymmetry, the guide describes an automorphism-group computation followed by a search for equal permutations with opposite signs. That is the basis of `EliminateDueSymmetries`.

## Practical Consequence

The architecture is optimized for many small comparison problems, which is typical in physics manipulations. The benchmark page then uses that same framing to explain why Redberry performs well on the tested workloads.
