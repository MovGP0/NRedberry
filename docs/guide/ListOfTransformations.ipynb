## Core Algebraic and Tensor Transformations

- `ApplyIndexMapping`: apply an explicit index mapping to a tensor.
- `Collect`, `CollectScalars`, `CollectNonScalars`: collect terms by a pattern or by scalar/tensorial structure.
- `Conjugate`, `Numeric`, `Numerator`, `Denominator`: basic scalar and rational manipulations.
- `Expand`, `ExpandAll`, `ExpandNumerator`, `ExpandDenominator`, `ExpandTensors`, `ExpandAndEliminate`: expansion helpers with different scopes.
- `Factor`, `Together`, `PowerExpand`, `PowerUnfold`: algebraic normalization and polynomial/rational rewrites.
- `EliminateMetrics`, `EliminateDueSymmetries`, `InvertIndices`: tensor-specific structural rewrites.
- `FullySymmetrize`, `FullyAntiSymmetrize`, `Symmetrize`: symmetry-producing transformations.
- `Identity`: the no-op transformation.
