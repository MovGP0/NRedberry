## Permutations

Permutations are a basic building block for tensor symmetries and many matching algorithms. The original guide shows both one-line and cycle notation, plus the sign flip that turns a symmetry into an antisymmetry.


```
var permutationNotations = new[]
{
    "[0, 2, 5, 6, 7, 1, 3, 4]",
    "[[1, 2, 5], [4, 7], [3, 6]]"
};

permutationNotations
```

```
var permutationActions = new[]
{
    "p >> [10, 9, 8, 7]",
    "p >> [\"a\", \"b\", \"c\", \"d\", \"e\"]",
    "perm1 * perm2"
};

permutationActions
```

## Permutation Groups

A generating set of permutations becomes a permutation group. The original guide uses the `Group(...)` constructor, then inspects the group order, enumerates elements, filters antisymmetries, tests membership, and computes stabilizers.


```
var groupQueries = new[]
{
    "group.order()",
    "group.membershipTest(perm)",
    "group.setwiseStabilizer(1, 3, 5)"
};

groupQueries
```

## Why It Matters

These group operations feed directly into tensor-symmetry handling, symmetrization, and mapping algorithms. They are not a side feature; they are part of the algebraic core of the system.
