## Preliminaries

The original guide explains Redberry through Groovy syntax. In the notebook port, the same ideas are expressed through C# and the NRedberry API.

Two core ideas appear throughout the documentation:

- creating expressions from string input
- applying transformations to expressions


### Expressions

Original Redberry/Groovy uses the `'...'.t` shorthand to parse a tensor expression. In NRedberry, use `Tensors.Parse(...)`.


```
var expr = TensorFactory.Parse("a * F^{A}_{B \\mu \\nu}");
expr
```

The input notation is very close to the Redberry/LaTeX-style syntax described in the original guide.

### Transformations

The original guide shows Groovy's `>>` operator for transformation application. In C#, use the transformation object directly.


```
var tensor = TensorFactory.Parse("(x + y) * a");
var expanded = NRedberry.Transformations.Expand.ExpandTransformation.Expand(tensor);
expanded
```

The exact convenience surface differs from Groovy Redberry, but the conceptual model is the same: transformations are first-class operations that act on symbolic expressions.

## Introductory Example

The original page walks through a scalar QED Compton-scattering example in Groovy. The main ideas are:

- represent vertices, propagators, and matrix elements symbolically
- apply substitutions
- expand and simplify
- use repeated substitutions for constraints such as on-shell conditions and momentum conservation

Below is a notebook-friendly C# sketch of that workflow. It is intentionally conservative and focuses on the symbolic manipulation pattern rather than claiming one-to-one API parity with the Groovy snippet.

```
var matrixElement = TensorFactory.Parse(
    "M^ij = V^i[k3_a,k3_a+k1_a]*D[k3_a+k1_a]*V^j[-k4_a,-k3_a-k1_a]"
    + "+V^j[k3_a,k3_a-k2_a]*D[k3_a-k2_a]*V^i[-k3_a+k2_a,-k4_a]"
    + "+V^ij");

matrixElement
```

This is the same overall structure as the original guide: build symbolic expressions first, then apply transformations and constraints in stages.

The original page also highlights that repetitive physics constraints can be applied programmatically. In notebook C#, the equivalent is an ordinary loop or any other standard C# control flow.

```
var shellConditions = new List<string>();
for (var i = 1; i <= 4; i++)
{
    shellConditions.Add($"k{i}_a*k{i}^a = 0");
}

shellConditions
```
