## Why Mappings Matter

In tensor algebra, comparison is not just a boolean equality check. Two expressions can be equivalent up to free-index renaming, dummy-index relabeling, and tensor symmetries. Redberry models the result of that comparison as an index mapping.


```polyglot-notebook
var lhs = TensorFactory.Parse("F_ab*G^bc");
var rhs = TensorFactory.Parse("F_iq*G^qj");

new
{
    Left = lhs.ToString(),
    Right = rhs.ToString(),
    ConceptualMapping = "{ a -> i, c -> j }"
}
```

## Multiple Mappings and Signs

When symmetries are present, a pair of tensors can admit several mappings. The original guide uses an antisymmetric tensor example where one mapping is positive and another is negative, representing a symmetry-versus-antisymmetry distinction.


```polyglot-notebook
var signedMappings = new[]
{
    "+{ a -> i, b -> j, c -> k }",
    "-{ a -> k, b -> j, c -> i }"
};

signedMappings
```

## Building Mappings Between Tensors

The original page uses the `%` operator to build a lazy mappings container and then iterate the available mappings on demand. That is the core mechanism behind substitutions and many matching-heavy algorithms.


```polyglot-notebook
var mappingWorkflow = new[]
{
    "from % to",
    "mappings.first",
    "foreach (var mapping in mappings) { ... }"
};

mappingWorkflow
```

## Constructing a Direct Mapping

The guide also constructs a mapping directly from explicit rules, for example `{ a -> b }` or `{ _a -> ^c, ^b -> _d }`, and then applies it to a tensor.


```polyglot-notebook
var mappedTensor = TensorFactory.Parse("f_a^b");

new
{
    Input = mappedTensor.ToString(),
    ConceptualRules = "{ _a -> ^c, ^b -> _d }",
    Expected = "f^c_d"
}
```

## Example Uses

The original page implements transformations like lowering all free upper indices or inverting free indices entirely by deriving a mapping from the tensor's own index structure and then applying it.


```polyglot-notebook
var sample = TensorFactory.Parse("f^ab_cd + t_mn*t^mnab_cd");

new
{
    Sample = sample.ToString(),
    LowerAllConcept = "(upperFree % invertedUpperFree) >> tensor",
    InvertConcept = "(from.si % to.si) >> tensor"
}
```
