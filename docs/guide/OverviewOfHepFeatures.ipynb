## High-Energy Physics Support

Beyond general tensor manipulation, Redberry includes a set of tools aimed at high-energy physics work. The original guide groups them into one-loop counterterms, equation solving, and common Dirac or SU(N) algebra utilities.


## One-Loop Counterterms

The original documentation highlights the curved-space one-loop counterterm tools for second- and fourth-order operators in four dimensions. In this repository, that material corresponds conceptually to the `NRedberry.Physics.Oneloopdiv` area and the related test suite.

## Solving Tensorial Equations

The original guide points to `Reduce` for converting tensorial equation systems into scalar systems that can be handed off to an external CAS. The main examples are propagator and projector constructions.
