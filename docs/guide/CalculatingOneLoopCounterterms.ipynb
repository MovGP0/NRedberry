## Background

The original page introduces Redberry's one-loop counterterm tools for curved-space calculations based on the extended t' Hooft-Veltman method as developed by Pronin and Stepanyantz.


## Operator Form

The central object is the differential operator derived from the second variation of the action. The page then defines the general higher-derivative form, the `Kn` and `(Kn)^(-1)` tensors, and the bundle curvature tensor required by the algorithm.


```polyglot-notebook
var operatorIngredients = new[]
{
    "K",
    "S",
    "W",
    "N",
    "M",
    "(Kn)^(-1)",
    "F_mu_nu"
};

operatorIngredients
```

## Supported Operator Families

The documented interface targets second-order and fourth-order operators. The current repository already reflects that focus in `NRedberry.Physics.Oneloopdiv` and the associated tests.


```polyglot-notebook
var supportedFamilies = new[]
{
    "D^(2) = K^{mu nu} nabla_mu nabla_nu + W",
    "D^(4) = K^{mu nu alpha beta} nabla_mu nabla_nu nabla_alpha nabla_beta + W^{mu nu} nabla_mu nabla_nu + M"
};

supportedFamilies
```

## Worked Example: Vector Field Operator

The page then gives a complete second-order example for a massive vector field. It defines the symmetries, `iK`, `K`, `S`, `W`, and `F`, runs `oneloopdiv2(...)`, and simplifies the resulting counterterms.


```polyglot-notebook
var vectorFieldWorkflow = new[]
{
    "addSymmetries \"R_abcd\"",
    "iK_a^b = d_a^b + g*n_a*n^b",
    "K^{mn}_i^j = ...",
    "S = 0",
    "W^a_b = P^a_b + g/(2*(1+g))*R^a_b",
    "F_abcd = R_abcd",
    "div = oneloopdiv2(iK, K, S, W, F)"
};

vectorFieldWorkflow
```

The original page also shows how to derive `iK` with `Reduce`, which is useful when it is not available in closed form at the outset.


```polyglot-notebook
var reduceUsage = new[]
{
    "K^ab_i^j*n_a*n_b * iK^i_m = d^j_m",
    "Reduce([...], [\"iK^a_b\"], options)"
};

reduceUsage
```
