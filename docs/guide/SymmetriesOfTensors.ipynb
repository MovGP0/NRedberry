## Specifying Symmetries

Redberry allows permutational symmetries to be attached to simple tensors and tensor fields. The classic example is the Riemann tensor, where only a small generating set is needed to recover the full permutation group.


```
// These calls mirror the original documentation syntax conceptually.
// Use the concrete symmetry-registration helpers exposed by the NRedberry port in your environment.

var riemannExpression = TensorFactory.Parse("R^abcd*R_efdc*R^ef_ab + R_rc^df*R_ab^rc*R_fd^ba");
riemannExpression
```

Once the relevant symmetries are registered, expressions that are equivalent up to those symmetries reduce automatically during parsing and standard-form normalization.


## Symmetries by Index Type

For tensors carrying more than one index type, the original guide shows how to attach symmetry only to a chosen type. That remains an important modeling pattern when Lorentz and internal indices coexist in the same symbol.


```
var mixedTypeTensor = TensorFactory.Parse("f_mn\\alpha\\beta + f_mn\\beta\\alpha");
mixedTypeTensor
```

## Permutation Groups

Internally, Redberry stores generators and works with a permutation-group representation. The original guide demonstrates querying the group order, stabilizers, and intersections after defining tensor symmetries.


```
var groupOperations = new[]
{
    "group.order()",
    "group.setwiseStabilizer(2, 3)",
    "group.intersection(otherGroup)"
};

groupOperations
```

## Inconsistent Generators

Symmetry generators can conflict, especially when antisymmetries are combined inconsistently. In those cases the permutation-group construction should fail rather than accept contradictory structure.


```
var inconsistentExample = new[]
{
    "[2, 3, 0, 1]",
    "-[1, 0, 2, 3]",
    "-[3, 2, 1, 0]"
};

inconsistentExample
```

## Finding Symmetries of Composite Tensors

The original guide also shows how to derive symmetries of a composite tensor expression from the symmetries of its ingredients. That workflow is useful when proving identities or reducing ansätze before solving equations.


```
var composite = TensorFactory.Parse("(R_abc*A_de + R_bde*A_ac)*A^ce + R_adb");
composite
```

## Multi-Term Symmetries

For Bianchi-like identities and Young-projector style reductions, the original guide uses an explicit substitution that encodes the multi-term symmetry. This notebook keeps that same approach: represent the projector relation as a symbolic substitution and then expand or simplify with it.


```
var youngProjectorSubstitution = "W_mnpq = 1/3*(2*W_mnpq - W_mqnp + W_mpnq)";
var weylIdentity = "W^p_q^r_s*W_p^t_r^u*W_tv^qw*W_u^vs_w";

new
{
    YoungProjectorSubstitution = youngProjectorSubstitution,
    WeylIdentity = weylIdentity
}
```
