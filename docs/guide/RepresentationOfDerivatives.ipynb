## Indexless Expressions

Redberry represents derivatives of pure functions with the `~( ... )` syntax. The notation records derivative order by argument position and then evaluates the resulting derivative at the supplied argument values.


```
var thirdDerivative = TensorFactory.Parse("f~(3)[x**2]");
var mixedDerivative = TensorFactory.Parse("f~(2, 3, 0)[a**2, w, q]");
var explicitDifferentiate = TensorFactory.Parse("D[x, y][y*x**2 + x*y**2 + f[x, y]]");

new
{
    ThirdDerivative = thirdDerivative.ToString(),
    MixedDerivative = mixedDerivative.ToString(),
    ExplicitDifferentiate = explicitDifferentiate.ToString()
}
```

The original documentation also demonstrates substitution into derivative expressions, for example replacing `f[x]` with `Sin[x]` and then evaluating a higher derivative.


```
var substitutionExample = new[]
{
    "f[x] = Sin[x]",
    "f~(3)[x**2]"
};

substitutionExample
```

## Indexed Expressions

For indexed tensor fields, the derivative-variable indices are appended to the tensor-field indices. The appended indices carry their own permutation symmetry because the order of differentiation is irrelevant within each argument slot.


```
var indexedDerivative = TensorFactory.Parse("F~(2)_mnab^cd[f_ab]");
var multiArgumentIndexedDerivative = TensorFactory.Parse("F~(1, 1)_mnabcde[f_abc, f_ab]");

new
{
    IndexedDerivative = indexedDerivative.ToString(),
    MultiArgumentIndexedDerivative = multiArgumentIndexedDerivative.ToString(),
    Indices = indexedDerivative.Indices.ToString()
}
```

The original guide emphasizes that appended derivative indices are assigned argument by argument. That means the placement reflects which argument was differentiated, not just the total number of appended slots.
