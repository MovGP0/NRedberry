## Standard Form

Redberry keeps expressions in a standard form during parsing and later transformations. This standardization is what makes matching, comparison, and term collection reliable across the system.


## Indexless Expressions

For scalar expressions, the standard form behaves like a conventional computer algebra system: terms are collected, common factors are reduced, and exact arithmetic is preserved where possible.


```
new
{
    SumReduction = TensorFactory.Parse("(a-b)+c+(b-a)").ToString(),
    ProductReduction = TensorFactory.Parse("c*(a-b)*(b-a)/c").ToString(),
    RationalReduction = TensorFactory.Parse("2/3-27**(1/3)/9").ToString(),
    NumericEvaluation = TensorFactory.Parse("Sin[2 + 2.*I]**(1/4)").ToString()
}
```

## Indexed Expressions

For indexed expressions, Redberry additionally tries to factor tensorial parts and keep scalar prefactors together. The distinction depends on the free-index structure of the terms being combined.


```
new
{
    FactoredTensorialPart = TensorFactory.Parse("a*F_mn + (a+b)*F_mn").ToString(),
    ZeroFreeIndexScalar = TensorFactory.Parse("(x_a^a + y_b^b)*X_m*X_n + (z_a^a - y_d^d)*X_m*X_n").ToString(),
    NonFactoredProduct = TensorFactory.Parse("x_a^a*X_m*X_n + y_a^a*X_m*X_n").ToString(),
    EquivalentTensorialPart = TensorFactory.Parse("(x_a^a + y_b^b)*X_m*X^m + (z_m^m - y_d^d)*X_a*X^a").ToString()
}
```

```
var scalarLike = TensorFactory.Parse("x_a^a + y_b^b");
var traced = TensorFactory.Parse("x_a^a");

new
{
    SumFreeIndexCount = scalarLike.Indices.Size(),
    TermIndexCount = traced.Indices.Size(),
    TermFreeIndexCount = traced.Indices.Free.Size()
}
```

## Ordering of Terms

The original guide notes that products and sums are ordered by internal hash codes. That ordering is stable within a session but can vary between runs unless the internal tensor-name generator seed is reset.


```
var product = TensorFactory.Parse("a*b*c");
product
```

## Internal Detail

Standard-form reduction is intentionally lightweight. It is not a replacement for heavier simplification passes, but it guarantees that expressions stay in a consistent canonical working shape throughout the system.
