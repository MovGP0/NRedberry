## First-Class Transformations

The original guide treats transformations as first-class objects: they can be assigned, combined, and applied repeatedly. In the notebook port, the same conceptual model applies even though the exact convenience syntax differs from the Groovy layer.


```polyglot-notebook
var input = TensorFactory.Parse("(A_k + B_k)*c");
var expanded = NRedberry.Transformations.Expand.ExpandTransformation.Expand(input);

new
{
    Input = input.ToString(),
    Expanded = expanded.ToString()
}
```

## Chaining Transformations

Redberry's guide uses `>>` and `<<` to apply transformations and `&` to join them into a single sequential transformation. Ordering matters: expand-then-substitute is not the same as substitute-then-expand.


```polyglot-notebook
var sequentialExamples = new[]
{
    "Expand & (c = a + b)",
    "(c = a + b) & Expand"
};

sequentialExamples
```

The original examples produce different outputs because transformation composition is ordered. The notebook version keeps that structure explicit so each stage stays readable.


## Parameterized Transformations

Some transformations take optional arguments, while others require arguments. The original page shows `Expand[EliminateMetrics]` and `Differentiate['x_m']` as representative examples.


```polyglot-notebook
var parameterizedExamples = new[]
{
    "Expand[EliminateMetrics]",
    "Differentiate[\"x_m\"]"
};

parameterizedExamples
```

```polyglot-notebook
var metricsExample = TensorFactory.Parse("(g_mn + d_m^a*g_na)*f^mn");
var differentiateExample = TensorFactory.Parse("x_m*x^m");

new
{
    MetricsExample = metricsExample.ToString(),
    DifferentiateExample = differentiateExample.ToString()
}
```

## Immutability

Transformations are immutable in the same way tensors are immutable. Creating a configured transformation yields a new object; it does not mutate a global transformation singleton.
