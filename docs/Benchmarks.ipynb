## Overview

The benchmark page compares Redberry against Cadabra, xAct, and Maple Physics on abstract tensor-simplification workloads. The framing is explicitly architectural: Redberry's graph-based comparison strategy is contrasted with canonicalization-heavy systems.


## Simplification Procedure

The original page documents the actual simplification routine used in the Redberry benchmarks, including repeated metric elimination, expansion, and symmetry elimination.


```
var redberryBenchmarkPipeline = new[]
{
    "EliminateMetrics",
    "d^i_i = 4",
    "Expand[EliminateMetrics & d^i_i = 4]",
    "EliminateDueSymmetries"
};

redberryBenchmarkPipeline
```

## Atomic Comparison Benchmark

The first benchmark studies comparison-like workloads by subtracting an expression from an equivalent rewritten form and measuring how long simplification takes to reach zero.


```
var atomicBenchmarkShape = new[]
{
    "sum of 100 products",
    "shuffle summands and multipliers",
    "rename dummies",
    "apply tensor symmetries",
    "simplify difference to zero"
};

atomicBenchmarkShape
```

## Deep Simplification Benchmark

The second benchmark uses nested sums of products of sums of products. The original conclusion is that Redberry gains a large advantage because it can simplify intermediate expressions aggressively during evaluation.


```
var deepBenchmarkShape = new[]
{
    "depth-4 expressions",
    "vary product sizes 2..4",
    "vary sum sizes 2..4",
    "compare Redberry, xAct, Cadabra"
};

deepBenchmarkShape
```
