# Redberry Documentation for .NET Interactive

This notebook is a .NET Interactive friendly version of the Redberry documentation landing page.

Source: [redberry.cc/documentation](http://redberry.cc/documentation)

The original site is written for Redberry's Java and Groovy APIs. This notebook is intended as a notebook-first index for working with the NRedberry port from C#.
